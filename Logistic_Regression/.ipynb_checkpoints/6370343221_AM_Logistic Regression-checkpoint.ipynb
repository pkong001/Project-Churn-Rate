## Logistic Regression

```python
# used for manipulating directory paths
import os

# Scientific and vector computation for python
import numpy as np

# Plotting library
from matplotlib import pyplot

# Optimization module in scipy
from scipy import optimize

# library written for this 
import utilsNN as utils

# tells matplotlib to embed plots within the notebook
%matplotlib inline
```

```python
# Load training dataset and test dataset

data_train = np.loadtxt(os.path.join('Data', 'Churn_Training_Data_mod.txt'))
data_cv = np.loadtxt(os.path.join('Data', 'Churn_CV_Data_mod.txt'))
data_test = np.loadtxt(os.path.join('Data', 'Churn_Test_Data_mod.txt'))

# First 10 columns of data are features and the last column is the label.
# Matrix X contains 10 features while vector y contains the label.
# Choose attributes of interest to analyze from the 10 features.
X_train, y_train = data_train[:, [0,1,3,5,6,7]], data_train[:, 10].astype(int)
X_cv, y_cv = data_cv[:, [0,1,3,5,6,7]], data_cv[:, 10].astype(int)
X_test, y_test = data_test[:, [0,1,3,5,6,7]], data_test[:, 10].astype(int)

print(X_train.shape)
print(y_train.shape)

m = y_train.size  # number of training examples
print(m)
```

```python
# Normalize code

def  featureNormalize(X):
    
    X_norm = X.copy()
    mu = np.zeros(X.shape[1])
    sigma = np.zeros(X.shape[1])

    mu = np.mean(X, axis=0)
    sigma = np.std(X, axis=0)
    X_norm = (X - mu) / sigma
    
    return X_norm, mu, sigma
```

```python
# Normalize training examples

print(type(X_train))

X_train, mu, sigma = featureNormalize(X_train)
X_cv, mu, sigma = featureNormalize(X_cv)
X_test, mu, sigma = featureNormalize(X_test)

print(type(X_train))
```

```python
# Sigmoid function code 

def sigmoid(z):
    z = np.array(z)
    
    g = np.zeros(z.shape)

    g = 1 / (1 + np.exp(-z))

    return g
```

```python
# Test the implementation of sigmoid function 

z = 0
g = sigmoid(z)

print('g(',z,') = ', g)
```

```python
# Setup the data matrix appropriately, and add ones for the intercept term

m_train, n_train = X_train.shape

# Add intercept term to X
X_train = np.concatenate([np.ones((m_train, 1)), X_train], axis=1)

m_cv, n_cv = X_cv.shape

# Add intercept term to X
X_cv = np.concatenate([np.ones((m_cv, 1)), X_cv], axis=1)

m_test, n_test = X_test.shape

# Add intercept term to X
X_test = np.concatenate([np.ones((m_test, 1)), X_test], axis=1)
```

```python
# Cost function code

def costFunction(theta, X, y):

    m = y.size  # number of training examples

    # You need to return the following variables correctly 
    J = 0
    grad = np.zeros(theta.shape)

    h = sigmoid(np.dot(X, theta))
    J = (1/m)*np.sum(-np.dot(y,np.log(h)) - np.dot((1-y),np.log(1-h)))
    grad = (1/m)*np.dot((h-y),X)
    
    return J, grad
```

```python
# Initialize fitting parameters

initial_theta = np.zeros(n_train+1)

print(initial_theta.shape)

cost, grad = costFunction(initial_theta, X_train, y_train)

print('Cost at initial zero theta parameter:\n', cost)
print('Gradient at initial zero theta parameter:\n', grad)
```

```python
# Learning parameters

options= {'maxiter': 200}

res = optimize.minimize(costFunction,
                        initial_theta,
                        (X_train, y_train),
                        jac=True,
                        method='TNC',
                        options=options)

cost = res.fun

# the optimized theta is in the x property
theta = res.x



print('Cost after gradient descent:\n', cost)
print('Theta after gradient descent:\n', theta)
```

```python
# Use logistic regression prediction model

def predict(theta, X):

    m = X.shape[0] # Number of training examples

    # You need to return the following variables correctly
    p = np.zeros(m)

    p=np.round(sigmoid(np.dot(X,theta)))
    
    return p
```

```python
# Compare data set accuracy by calling prediction function

p_train = predict(theta, X_train)
print('Training Accuracy: {:.2f} %'.format(np.mean(p_train == y_train) * 100))

p_cv = predict(theta, X_cv)
print('Cross Validation Accuracy: {:.2f} %'.format(np.mean(p_cv == y_cv) * 100))

p_test = predict(theta, X_test)
print('Test Accuracy: {:.2f} %'.format(np.mean(p_test == y_test) * 100))
```

## Regularized Logistic Regression

```python
# Regularized cost function code

def costFunctionReg(theta, X, y, lambda_=1):

    m = y.size  # number of training examples

    # You need to return the following variables correctly 
    J = 0
    grad = np.zeros(theta.shape)
    th = theta[1:]

    h = sigmoid(np.dot(X, theta))
    J = J = (1/m)*np.sum(-np.dot(y,np.log(h)) - np.dot((1-y),np.log(1-h))) + ( lambda_ / (2 * m) )* (np.sum((np.power(th,2))))
    
    grad_no_regulation = (1/m)*np.dot((h-y),X)
    grad = (1/m)*np.dot((h-y),X) + (lambda_ / m) * theta
    
    grad[0] = grad_no_regulation[0]

    return J, grad  
```

```python
# Initialize fitting parameters

initial_theta = np.zeros(X_train.shape[1])
lambda_ = 1

cost, grad = costFunctionReg(initial_theta, X_train, y_train, lambda_)

print('Cost at initial zero theta parameter:\n', cost)
print('Gradient at initial zero theta parameter:\n', grad)
```

```python
# Learning parameters

options= {'maxiter': 200}

res = optimize.minimize(costFunctionReg,
                        initial_theta,
                        (X_train, y_train),
                        jac=True,
                        method='TNC',
                        options=options)

cost = res.fun

# the optimized theta is in the x property
theta = res.x

print('Cost after gradient descent:\n', cost)
print('Theta after gradient descent:\n', theta)
```

```python
# Compare data set accuracy by calling prediction function

p_train = predict(theta, X_train)
print('Training Accuracy: {:.2f} %'.format(np.mean(p_train == y_train) * 100))

p_cv = predict(theta, X_cv)
print('Cross Validation Accuracy: {:.2f} %'.format(np.mean(p_cv == y_cv) * 100))

p_test = predict(theta, X_test)
print('Test Accuracy: {:.2f} %'.format(np.mean(p_test == y_test) * 100))
```

## Learning Curve

```python
# Learning curve code

def learningCurve(X, y, Xval, yval, lambda_=0):

    # Number of training examples
    m = y.size

    # You need to return these values correctly
    error_train = np.zeros(m)
    error_val   = np.zeros(m)
            
    for i in range(1, m+1):
        Xtrain = X[:i, :];
        ytrain = y[:i];
        res = optimize.minimize(costFunctionReg,initial_theta,(Xtrain, ytrain, lambda_),jac=True,method='TNC',options=options)
        theta = res.x
        
        error_train[i-1] = costFunctionReg(theta, Xtrain, ytrain, lambda_)[0]
        error_val[i-1] = costFunctionReg(theta, Xval, yval, lambda_)[0]
        
    return error_train, error_val
```

```python
# Plot learning curve

X_aug = X_train
Xval_aug = X_cv

error_train, error_val = learningCurve(X_aug, y_train, Xval_aug, y_cv, lambda_= 0)

pyplot.plot(np.arange(1, m+1), error_train, np.arange(1, m+1), error_val, lw=2)
pyplot.title('Learning curve for logistic regression')
pyplot.legend(['Train', 'Cross Validation'])
pyplot.xlabel('Number of training examples')
pyplot.ylabel('Error')
pyplot.axis([0, 2000, 0, 1])

print('# Training Examples\tTrain Error\tCross Validation Error')
for i in range(m):
    print('  \t%d\t\t%f\t%f' % (i+1, error_train[i], error_val[i]))
```

```python
# Output error for each regularization parameter

lambda_array = np.array([0.01, 0.03, 0.1, 0.3, 1, 3, 10, 30])

for i in range(lambda_array.size):
    res = optimize.minimize(costFunctionReg,initial_theta,(X_train, y_train, lambda_array[i]),jac=True,method='TNC',options=options)
    theta = res.x
    error_train = costFunctionReg(theta, X_train, y_train, lambda_)[0] 
    error_cv = costFunctionReg(theta, X_cv, y_cv, lambda_)[0]
    print('Lambda value: ',lambda_array[i])
    print('Training error: ',error_train)
    print('Cross validation error: ',error_cv)
```
